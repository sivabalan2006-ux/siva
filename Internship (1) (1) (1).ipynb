```python
import matplotlib.pyplot as plt


df['status']=df['Grade'].apply(lambda x:'pass' if x !='F'else 'Fail')

df['status'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title('Pass vs Fail')
plt.ylabel('')

plt.show()
```
